```python
missing_values = df.isnull().sum()

# Show the columns with missing values and their count
print("\nMissing Values per Column:")
print(missing_values)
```

```python
# Separate numerical and categorical columns
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Further clean categorical columns to ensure they contain only non-numeric data
categorical_cols = [col for col in categorical_cols if df[col].apply(lambda x: isinstance(x, str)).all()]

# Impute missing values for numerical columns with mean
df[numerical_cols] = df[numerical_cols].fillna(df[numerical_cols].mean())

# Impute missing values for categorical columns with mode
df[categorical_cols] = df[categorical_cols].apply(lambda col: col.fillna(col.mode().iloc[0]) if col.dtype == 'object' else col)

# Show the result of missing value imputation
print("\nMissing Values After Imputation:")
print(df.isnull().sum())
```

```python
# Check the column names to verify the correct name
print(df.columns)

# Clean column names (remove leading/trailing spaces)
df.columns = df.columns.str.strip()

# After cleaning, check the columns again
print(df.columns)
```
